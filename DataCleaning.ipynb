## Changing NaN values to median of their respective columns

```python
import pandas as pd
import numpy as np

# Assuming productsreviews_cleaned is your cleaned dataframe
# Define numeric columns to consider for NaN replacement
numeric_columns = productsreviews_cleaned.select_dtypes(include=['int64', 'float64']).columns

# Exclude non-numeric columns that should not be filled
columns_to_exclude = ['product_id', 'product_name', 'brand_name', 
                      'primary_category', 'secondary_category', 
                      'tertiary_category']

# Identify columns to fill NaN values
columns_to_fill = [col for col in numeric_columns if col not in columns_to_exclude]

# Fill NaN values with median of their respective columns
productsreviews_cleaned_filled = productsreviews_cleaned.copy()
productsreviews_cleaned_filled[columns_to_fill] = productsreviews_cleaned[columns_to_fill].fillna(productsreviews_cleaned[columns_to_fill].median())

# Check for duplicates
duplicate_rows = productsreviews_cleaned_filled[productsreviews_cleaned_filled.duplicated()]

# Print duplicate rows if any
if not duplicate_rows.empty:
    print("Duplicate rows found:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")

# Display the cleaned and filled dataframe
print("Shape of cleaned dataframe:", productsreviews_cleaned_filled.shape)
print(productsreviews_cleaned_filled.head())
```
