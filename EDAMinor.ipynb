```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
from sqlalchemy import create_engine


plt.style.use('ggplot')

ProductsReviewsCC = pd.read_csv('productsreviews_cleaned_filled.csv')

```

```python
# Displaying primary product category

plt.figure(figsize=(12, 6))

ProductsReviewsCC['primary_category'].value_counts().iloc[::-1].plot(kind='barh', color='maroon', edgecolor='black')

plt.title('Primary Product Category')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Displaying Secondary product categories

plt.figure(figsize=(12, 12))

ProductsReviewsCC['secondary_category'].value_counts().iloc[::-1].plot(kind='barh', color='lightcoral', edgecolor='dimgrey')

plt.title('Secondary Product Category')
plt.xlabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Displaying Tertiary Category. Category used for recommendations.

plt.figure(figsize=(12, 12))

ProductsReviewsCC['tertiary_category'].value_counts().iloc[::-1].plot(kind='barh', color='lightcoral', edgecolor='dimgrey')

plt.title('Tertiary Category')
plt.xlabel('Count')

plt.tight_layout()
plt.show()
```
